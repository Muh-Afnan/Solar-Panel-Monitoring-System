```python
import os
from PIL import Image

# List of dataset folders
image_folders = [
    r"D:\Machine Learning Projects\Solar Panel Monitoring System\Data\Raw\Dirty",
    r"D:\Machine Learning Projects\Solar Panel Monitoring System\Data\Raw\Clean",
    r"D:\Machine Learning Projects\Solar Panel Monitoring System\Data\Snow-Covered",
    r"D:\Machine Learning Projects\Solar Panel Monitoring System\Data\Physical-damage",
    r"D:\Machine Learning Projects\Solar Panel Monitoring System\Data\Electrical-damage"
]

# Supported image formats
supported_extensions = (".jpg", ".jpeg", ".png")

# Loop through each folder
for folder in image_folders:
    image_files = [file for file in os.listdir(folder) if file.lower().endswith(supported_extensions)]
    print(f"\n‚úÖ Folder: {os.path.basename(folder)}")
    print(f"üîç Total Images Found: {len(image_files)}")

    valid_sizes = []
    rejected_sizes = []

    for img_name in image_files:
        img_path = os.path.join(folder, img_name)

        try:
            img = Image.open(img_path)
            width, height = img.size

            # Aspect ratio filter (adjustable threshold)
            aspect_ratio = width / height

            if 0.65 <= aspect_ratio <= 1.40:
                valid_sizes.append((width, height))
            else:
                rejected_sizes.append((width, height))

        except Exception as e:
            print(f"‚ö†Ô∏è Error opening {img_path}: {e}")

    print(f"‚úÖ Valid Images: {len(valid_sizes)}")
    print(f"üìè Invalid Images: {len(rejected_sizes)}")
    print(f"üìè Invalid Images Percentage: {(len(rejected_sizes)/len(image_files))*100}")


```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img, array_to_img

# ‚úÖ Folder where your original images are stored (e.g., Clean, Dirty, etc.)
input_folder = r"D:\Machine Learning Projects\Solar Panel Monitoring System\Data\Physical-damage"

# ‚úÖ Folder to save augmented images
output_folder = r"D:\Machine Learning Projects\Solar Panel Monitoring System\Data\Electrical-damage"
os.makedirs(output_folder, exist_ok=True)

# ‚úÖ ImageDataGenerator for augmentation
datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# ‚úÖ How many augmentations per image
augmentations_per_image = 3

# ‚úÖ Process each image
for img_name in os.listdir(input_folder):
    if img_name.lower().endswith(('.jpg', '.jpeg', '.png')):
        img_path = os.path.join(input_folder, img_name)

        try:
            # Load image and convert to numpy array
            img = load_img(img_path)
            x = img_to_array(img)
            x = x.reshape((1,) + x.shape)

            # Create augmentations
            i = 0
            for batch in datagen.flow(x, batch_size=1):
                aug_img = array_to_img(batch[0])
                new_name = f"{os.path.splitext(img_name)[0]}_aug_{i+1}.png"
                aug_img.save(os.path.join(output_folder, new_name))
                i += 1
                if i >= augmentations_per_image:
                    break

        except Exception as e:
            print(f"Error processing {img_name}: {e}")
```
